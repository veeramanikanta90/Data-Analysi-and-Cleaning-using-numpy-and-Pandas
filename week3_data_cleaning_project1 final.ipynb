```python
dataset['Easy Apply'].replace("-1",False,inplace=True)
dataset.head(5)
```

```python
dataset['Established'].replace(-1,1999,inplace=True)
dataset.head(5)
```

```python
dataset['Rating'].replace(-1.0,0.0,inplace=True)
dataset.head()
```

```python
dataset['Age']=dataset['Age'].astype(int)
```

```python
dataset.head(5)
```

```python
del dataset[dataset.columns[0]]
```

```python
dataset.head(5)
```

```python
dataset['Salary']=dataset['Salary'].apply(lambda dataset:dataset.replace("$",''))
dataset.head()
```

```python
dataset['Salary']=dataset['Salary'].apply(lambda dataset:dataset.replace("k",'000'))
dataset.head()
```

```python
dataset['Location']=dataset['Location'].str.replace('Australia Aus','Australia,Aus')
dataset['Location']=dataset['Location'].str.replace('India In','India,In')
```
